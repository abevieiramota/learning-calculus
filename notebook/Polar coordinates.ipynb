```python
from matplotlib import pyplot as plt
%matplotlib inline

import numpy as np
```

# circle centered at (a, 0) with radius a

```python
a = 2

# from -pi/2 to pi/2
theta = np.linspace(-1 * np.pi / 2., np.pi / 2., 100)
r = 2 * a * np.cos(theta)
```

```python
plt.figure(figsize=(15, 5))
ax1 = plt.subplot(131, projection='polar')
ax2 = plt.subplot(133)

ax1.plot(theta, r)

for thetai in theta:
    
    # x point coordinate
    xi = 2 * a * np.cos(thetai)**2
    # 10 eqspaced values from 0 to xi
    xr = np.linspace(0, xi, 10)
    # y(line) points coordinates
    yr = np.tan(thetai) * xr
    
    ax2.plot(xr, yr, c='c')
    
plt.gca().set_aspect('equal')
```

# 4-leaf rose

```python
theta = np.linspace(0, np.pi * 2, 200)
r = np.sin(2* theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')

ax.plot(theta, r)
```
