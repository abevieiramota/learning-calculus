```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

# Contour

```python
z = np.meshgrid(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = x

xx,yy = np.meshgrid(x, y)
zz = xx * yy

plt.contourf(xx, yy, zz, cmap=plt.cm.viridis)
plt.colorbar()
```

```python
x = np.linspace(-10, 10, 100)
y = x

xx,yy = np.meshgrid(x, y)
zz = xx + yy

plt.contourf(xx, yy, zz, cmap=plt.cm.viridis)
plt.colorbar()
```

```python
x = np.linspace(-10, 10, 100)
y = x

xx,yy = np.meshgrid(x, y)
zz = xx**2 + yy**2

plt.contourf(xx, yy, zz, cmap=plt.cm.viridis)
plt.colorbar()
```
