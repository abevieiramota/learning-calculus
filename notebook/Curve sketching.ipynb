```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
a = 10
b = 4
c = 23
x = np.linspace(-10, 10, 1000)
y1 = b*x + c
y2 = a*x**2
y3 = y1 + y2

plt.figure(figsize=(10, 10))
plt.plot(x, y1, label='$bx + c$')
plt.plot(x, y2, label='$ax^2$')
plt.plot(x, y3, label='$ax^2 + bx + c$')

plt.legend()
```
